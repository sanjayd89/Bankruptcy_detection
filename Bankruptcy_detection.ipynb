## 3.1 Outlier Removal from Categorical Columns

- Keeping both categorical columns at end to facilitate analysis by swapping the position of ' Liability-Assets Flag' with ' Equity to Liability'

```python
#define function to swap columns
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df

#swap points and rebounds columns
data = swap_columns(data, 'Liability-Assets Flag', 'Equity to Liability')
```

## 3.2 Function to calculate Upper & Lower limit using IQR

```python
outlier_count__data_percent = []
for col_name in X_columns:
  q1, q3 = np.percentile(data[col_name], 25), np.percentile(data[col_name], 75)
  IQR = q3 - q1
  upper_limit = q3 + (1.5 * IQR)
  lower_limit = q1 - (1.5 * IQR)
  outlier = [x for x in data[col_name] if x < lower_limit or x > upper_limit]
  outlier_count_percent = round((len(outlier)/data.shape[0])*100,2)   
  outlier_count__data_percent.append(outlier_count_percent)
```

```python
dictionary = {
    'feature_name' : X_columns,
    'outlier_percent' : outlier_count__data_percent
}

outlier_data = pd.DataFrame(dictionary)
```

```python
outlier_data.set_index('feature_name').sort_values('outlier_percent', ascending = True).plot(kind = 'barh', figsize = (15, 25))
```

```python
print('No. of columns with more than 20% outlier: ',outlier_data[outlier_data['outlier_percent']>20].shape[0])
```

- 5 columns having number of outlier % more than 20 will be dropped as imputing these values may create bias in the model which might negatively affect the model
- Rest of the columns, the outliers will be replaced with upper & lower limit values calculated from IQR

```python
columns_to_drop = outlier_data[outlier_data['outlier_percent']>20]['feature_name']
print('before dropping data shape: ', data.shape)
data = data.drop(columns_to_drop, axis = 1)
print('after dropping data shape: ', data.shape)
X_columns = data.columns[1:data.columns.shape[0]-1]
```

```python
data.describe().transpose()['min'].min()
```

- minimum value of all columns does not go below zero. Thus, while imputing for outliers, the lower_limit shall be max(lower_limit, 0)

## 3.3 Outlier Removal

```python
#removing outliers from remaining columns and replacing with median i.e. 50th percentile

for col in X_columns:  
  q1, q2, q3 = np.percentile(data[col], 25), np.percentile(data[col], 50), np.percentile(data[col], 75)
  IQR = q3 - q1
  upper_limit = q3 + (1.5 * IQR)
  lower_limit = q1 - (1.5 * IQR)
  data.loc[data[col]< lower_limit, col] = q2
  data.loc[data[col]> upper_limit, col] = q2  
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```
